```python
import numpy  as np
import pandas as pd
from sklearn import tree
from sklearn.cluster import KMeans
```

```python
csv = pd.read_csv('CASAS.csv')
df_living = pd.read_csv('CASAS.csv', usecols=['sqft_living'])
df_lot = pd.read_csv('CASAS.csv', usecols=['sqft_lot'])
df_above = pd.read_csv('CASAS.csv', usecols=['sqft_above'])
df_basement = pd.read_csv('CASAS.csv', usecols=['sqft_basement'])
```

```python
X_living = np.array(df_living)
X_lot = np.array(df_lot)
X_above = np.array(df_above)
X_basement = np.array(df_basement)
```

```python
X_lot
```

```python
k_living = KMeans(n_clusters=3, random_state=0)
k_lot = KMeans(n_clusters=3, random_state=0)
k_above = KMeans(n_clusters=3, random_state=0)
k_basement = KMeans(n_clusters=3, random_state=0)
```

```python
k_living.fit(X_living)
k_lot.fit(X_lot)
k_above.fit(X_above)
k_basement.fit(X_basement)
```

```python
k_above.labels_
```

```python
k_living.labels_
```

```python
csv['K-sqft_living'] = k_living.labels_
csv['K-sqft_lot'] = k_lot.labels_
csv['K-sqft_above'] = k_above.labels_
csv['K-sqft_basement'] = k_basement.labels_
```
